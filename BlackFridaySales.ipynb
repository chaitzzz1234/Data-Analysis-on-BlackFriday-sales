```python
df = df.drop(['Product_Category_2', 'Product_Category_3'], axis=1)
```

```python
df.isnull().sum()
```

**GENDER ANALYSIS**

```python
df['Gender'].value_counts()
```

```python
df['Gender'].value_counts().plot(kind='bar',figsize=(5,5),color='red')
```

```python
df['Gender'].value_counts().plot(kind='pie',figsize=(5,5),color='red')
```

**GENDER ANALYSIS ACCORDING TO  PURCHASES**

```python
df.groupby('Gender')['Purchase'].sum().plot(kind='bar', figsize=(5, 5))
```

```python
df.groupby('Gender')['Purchase'].sum().plot(kind='pie', figsize=(5, 5))
```

**AGE & MARTIAL STATUS ANALYSIS**
